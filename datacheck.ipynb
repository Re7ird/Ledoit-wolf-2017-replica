```python
def openfile(filename):
    a=open(filename,'r')
    b=a.readlines()
    a.close()
    return b

stock=openfile("S&P500.csv")
stocklist=stock[0].replace("\n","").split(",")
print(stocklist)
print("length:"+str(len(stocklist)))
```
