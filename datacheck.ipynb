```python
def openfile(filename):
    a=open(filename,'r')
    b=a.readlines()
    a.close()
    return b

stock=openfile("S&P500.csv")
a=stock[0].replace("\n","").split(",")
stocklist=a

stock=openfile("listofchanges1.csv")
a=stock[0].replace("\n","").split(",")
stocklist=stocklist+a
#print(stocklist)
print("length:"+str(len(stocklist)))
```
